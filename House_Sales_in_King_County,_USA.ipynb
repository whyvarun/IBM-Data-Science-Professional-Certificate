```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.model_selection import train_test_split, cross_val_score

# Ensure graphs display inline in Jupyter
%matplotlib inline

# Load the dataset
file_name = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/FinalModule_Coursera/data/kc_house_data_NaN.csv'
df = pd.read_csv(file_name)

# --- Data Cleaning (From Module 2) ---
# Replace missing 'bedrooms' values with the mean
mean_bedrooms = df['bedrooms'].mean()
df['bedrooms'].replace(np.nan, mean_bedrooms, inplace=True)

# Replace missing 'bathrooms' values with the mean
mean_bathrooms = df['bathrooms'].mean()
df['bathrooms'].replace(np.nan, mean_bathrooms, inplace=True)

print("Data loaded and cleaned.")
```

```python
# Question 1: Display the data types of each column
print(df.dtypes)
```

```python
# Question 2: Drop columns and describe
df.drop(['id', 'Unnamed: 0'], axis=1, inplace=True)
df.describe()
```

```python
# Question 3: Value counts for floors
floor_counts = df['floors'].value_counts().to_frame()
print(floor_counts)
```

```python
# Question 4: Boxplot of price vs waterfront
sns.boxplot(x='waterfront', y='price', data=df)
plt.title("Price distribution for Waterfront vs Non-Waterfront")
plt.show()
```

```python
# Question 5: Regplot for sqft_above vs price
sns.regplot(x='sqft_above', y='price', data=df)
plt.title("Regression Plot: sqft_above vs Price")
plt.ylim(0,)
plt.show()
```

```python
# Question 6: Linear Regression (sqft_living)
X = df[['sqft_living']]
Y = df['price']
lm = LinearRegression()
lm.fit(X, Y)
print("R^2 for sqft_living:", lm.score(X, Y))
```

```python
# Question 7: Linear Regression (multiple features)
features = ["floors", "waterfront", "lat", "bedrooms", "sqft_basement", "view", "bathrooms", "sqft_living15", "sqft_above", "grade", "sqft_living"]
Z = df[features]
lm.fit(Z, df['price'])
print("R^2 for multiple features:", lm.score(Z, df['price']))
```

```python
# Question 8: Pipeline
Input = [('scale', StandardScaler()), ('polynomial', PolynomialFeatures(include_bias=False)), ('model', LinearRegression())]
pipe = Pipeline(Input)
pipe.fit(Z, df['price'])
print("R^2 for Pipeline:", pipe.score(Z, df['price']))
```

```python
# --- Data Splitting for Q9 & Q10 ---
features = ["floors", "waterfront", "lat", "bedrooms", "sqft_basement", "view", "bathrooms", "sqft_living15", "sqft_above", "grade", "sqft_living"]
X = df[features]
Y = df['price']
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.15, random_state=1)

# Question 9: Ridge Regression
RidgeModel = Ridge(alpha=0.1)
RidgeModel.fit(x_train, y_train)
print("R^2 for Ridge (test data):", RidgeModel.score(x_test, y_test))
```

```python
# Question 10: Polynomial Transform + Ridge Regression
pr = PolynomialFeatures(degree=2)
x_train_pr = pr.fit_transform(x_train)
x_test_pr = pr.fit_transform(x_test)

RidgeModel = Ridge(alpha=0.1)
RidgeModel.fit(x_train_pr, y_train)
print("R^2 for Polynomial Ridge (test data):", RidgeModel.score(x_test_pr, y_test))
```
